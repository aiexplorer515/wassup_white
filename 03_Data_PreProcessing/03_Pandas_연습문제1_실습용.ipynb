# Q1-1

```
아래의 데이타를 이용하여 데이타프레임을 생성하여라.
name_list는 인덱스로 이용한다. 

data = {'kor':[40, 55, 66, 77, 34],
        'math':[70, 55, 66, 20, 40],
        'eng':[80, 95, 100, 77, 90]}
        
name_list = ['고길동','김희동','신윤지','이민호','박준희']        
```

````python
#Q1-1
data = {'kor':[40, 55, 66, 77, 34],
        'math':[70, 55, 66, 20, 40],
        'eng':[80, 95, 100, 77, 90]}
        
name_list = ['고길동','김희동','신윤지','이민호','박준희'] 
df1 = pd.DataFrame(data, index=name_list)
df1
````

# Q1-2
```
2명의 데이타를 추가하여라. 

신머신	55	67	88
윤파이	90	100	90
```

````python
new_data = {'kor':[55,90],
        'math':[67,100],
        'eng':[88,90]}
        
name_list1 = ['신머신','윤파이'] 
df1 = pd.concat([df1, pd.DataFrame(new_data, index=name_list1)])
df1
````

# Q1-3
sci 과목의 데이타를 추가하여라 

90, 45, 77, 90, 67, 90, 80

````python
df1['sci']=[90, 45, 77, 90, 67, 90, 80]
df1
````

# Q1-4
eng 과목에서 eng 평균점수보다 높은 학생은?

````python
df2= df1
eng_mean=round(df2['eng'].mean(),2)
df2.loc[df2.eng >= eng_mean]
````

# Q1-5
```
전체 과목의 평균을 구한 후 평균이상이면 O, 평균미만이면 X를 pass 컬럼에 출력하여라.
```

````python
df2['평균']=df2.mean(axis=1)
df_mean = round(df2['평균'].mean(), 2)
df2['pass']=np.where(df2['평균']>df_mean, 'O','X')
df2
````
